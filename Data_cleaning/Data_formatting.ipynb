```python
# Assuming your dataframe is named article_df and has a column 'title' like "Article 14"
# First, extract the article number
articles_df["article_number"] = (
    articles_df["title"]
    .str.extract(r"(\d+)")  # extract digits
    .astype(float)          # convert to number
)

# Define the mapping of Parts to their Article ranges
parts_map = {
    "Part I": (1, 4),
    "Part II": (5, 11),
    "Part III": (12, 35),
    "Part IV": (36, 51),
    "Part V": (52, 151),
    "Part XII": (264, 300),
    "Part XIII": (301, 307)
}

# Function to assign part based on article number
def assign_part(num):
    for part, (start, end) in parts_map.items():
        if start <= num <= end:
            return part
    return "Other"

# Apply mapping
articles_df["Part"] = articles_df["article_number"].apply(assign_part)

# Drop helper column if not needed
articles_df.drop(columns=["article_number"], inplace=True)

# Save updated dataset
articles_df.to_csv("Articles_with_parts.csv", index=False)
```
