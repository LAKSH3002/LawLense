```python
# Now for each query, we will extract results and their respective pages, so there are going to be at least 6200 cases in total
base_url = "https://indiankanoon.org/search/?formInput="
first_50_data = []
second_50_data = []
third_50_data = []
last_50_data = []
miscellaneous_data = []
```

```python
for query in first_50_queries_articles_results:
    search_url = base_url + query.replace(" ", "+")
    response = requests.get(search_url)
    soup = BeautifulSoup(response.text, "html.parser")

    results = soup.find_all("div", class_="result_title")

    for r in results[:10]:  # top 10 per query
        link_tag = r.find("a")
        if link_tag:
            link = "https://indiankanoon.org" + link_tag["href"]
            title = r.text.strip()

            first_50_data.append({
                "query": query,
                "case_title": title,
                "case_url": link
            })
    
    print(f"✅ Collected {len(results[:25])} results for '{query}'")
    time.sleep(random.uniform(3, 5))
```
