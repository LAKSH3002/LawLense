```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# The format of the url for articles of the constitution on the Indian Kanoon website after searching about article is
# https://indiankanoon.org/search/?formInput=Article+1
# https://indiankanoon.org/search/?formInput=Article+2

# Thus just the number in the end needs to be replaced all the time

base_url = "https://indiankanoon.org/search/?formInput=Article+{}"

# Empty list to gather data in dictionary format
data = []
data2 = []
```

```python
# Plan - Scraping 195 articles (1-151 & 264-307). 
# These articles make up 7 parts of the constitution are the most relevant to public, finance and small businesses
# Part description along with the articles
# Part 1: The Union and its territories: 1-4
# Part 2: Citizenship: 5-11
# Part 3: Fundamental rights: 12-35
# Part 4: Directive principles of state policy: 36-51
# Part 5: The Union: 52-151
# Part 12: Finance, property, Contracts and Suits: 264-300A
# Part 13: Trade, commerce and inter course within the territory of India: 301-307
```

```python
# Running a for loop in order to first extract 
# div class = "result", div class = "result_title"
for i in range(1, 152):  # Fetch Articles 14 to 64
    search_url = base_url.format(i)
    # Obtaining response
    response = requests.get(search_url)
    # Obtaining html content
    soup = BeautifulSoup(response.text, 'html.parser')

    # Obtaining all results of all the links that will be provided
    results = soup.find_all('div', class_='result')

    found = False
    for r in results:
        # Extracting the result_title
        title = (
            r.find('div', class_='result_title').text.strip()
            if r.find('div', class_='result_title')
            else ""
        )
        # Generating the link and navigating to the link
        link = (
            "https://indiankanoon.org" + r.find('a')['href']
            if r.find('a')
            else ""
        )

        # https://indiankanoon.org/search/?formInput=Article+33 - Article search link
        # https://indiankanoon.org/doc/237570/ - article document link

        # Going to that article link and extracting content
        # Ensuring both 'Article {i}' and 'in Constitution of India' appear in the data.
        if f"Article {i}" in title and "in Constitution of India" in title:
            if link:
                article_page = requests.get(link)
                soup2 = BeautifulSoup(article_page.text, 'html.parser')
                content_div = soup2.find('section', class_='akn-section')
                content = content_div.get_text(separator=" ", strip=True) if content_div else ""
            else:
                content = ""

            data.append({
                "url": link,
                "title": title,
                "content": content
            })
            found = True
            print(f"✅ Collected Article {i}")
            break

    if not found:
        print(f"⚠️ No valid match for Article {i}")
    time.sleep(1)
```
